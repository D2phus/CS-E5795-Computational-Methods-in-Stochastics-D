## Definitions

```python
import random
import matplotlib.pyplot as plt
import numpy as np
n_interactions = 100
n_interactions_c = 10
n_agents = 500000
moods_a = [50]*n_agents
moods_b = [50]*n_agents
moods_c = [50]*n_agents
```

# Part a) generation

```python
for agent in range(n_agents):
    for interaction in range(n_interactions):
        rand = random.uniform(0, 1)
        if rand > 0.5:
            moods_a[agent] += 1
        else:
            moods_a[agent] -= 1
            
```

# Part b) generation

```python
for agent in range(n_agents):
    for interaction in range(n_interactions):
        rand = random.uniform(0, 1)
        if rand > 0.5:
            moods_b[agent] += 10
        else:
            moods_b[agent] -= 0.5
```

# Part c) generation

```python
for agent in range(n_agents):
    for interaction in range(n_interactions_c):
        rand = random.uniform(0, 1)
        if rand > 0.5:
            moods_c[agent] *= 1/0.7
        else:
            moods_c[agent] *= 0.7
```

# Plot a) and b)
a) Appears to be a binomial distribution (PMF) which approaches a normal density function (PDF)

b) Also apperars to be a binomial distribution (PMF) which approaches normal density function (PDF) 


```python
plt.hist(moods_a, bins=100)
plt.show() 
```

```python
plt.hist(moods_b, bins=100)
plt.show() 
```
