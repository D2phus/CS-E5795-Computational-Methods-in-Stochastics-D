## 1.

```python
x_i = 1

def ggl_set(number):
    global x_i
    x_i = number
    
def ggl_get():
    return x_i

def ggl():
    a = 16807
    m = 2**31 - 1 
    x_j = (a * ggl_get()) % m 
    ggl_set(x_j)
    return x_j / m

ggl_set(1)

for i in range(10):
    res = ggl()
    print(res)
```

```python
arr = []

def ran3_init():
    global arr
    ggl_set(1)
    arr = [ggl() for i in range(55)] 
    
def ran3_get():
    x_i55 = arr[0]
    x_i24 = arr[31]
    return x_i55, x_i24

def ran3_set(number):
    global arr
    arr.pop(0)
    arr.append(number)

def ran3():
    x_j55, x_j24 = ran3_get()
    m = 2**32
    x_j = (x_j55 - x_j24) % m
    ran3_set(x_j)
    return x_j / m 

ran3_init()

for i in range(10):
    res = ran3()
    print(res)
```

```python
import matplotlib.pyplot as plt
import random as rand

def plot(x, y):
    plt.plot(x, y, 'bo')
    
    plt.ylabel('y')
    plt.xlabel('x')
    
    plt.xlim(0, 10**(-4))
    plt.ylim(0, 1)
    
    plt.show()
    
def ggl_plot():
    x = []
    y = []
    
    ggl_set(1)
    for i in range(10000000):
        x.append(ggl())
        y.append(ggl())
    
    plt.title("GGL")
    plot(x, y)
    
def ran3_plot():
    x = []
    y = []
    
    ran3_init()
    for i in range(10000000):
        x.append(ran3())
        y.append(ran3())
     
    plt.title("RAN3")
    plot(x, y)
    
def random_plot():
    x = []
    y = []
    
    rand.seed(1)
    for i in range(10000000):
        x.append(rand.random())
        y.append(rand.random())
    
    plt.title("Random")
    plot(x, y)


ggl_plot()
ran3_plot()
random_plot()
```

#### Findings
When looking at the plots the built-in random generator in Python seems to be most random. The points are pretty evenly scattered. There are some blank spots and some spots with more points but they still seem random.

The plot for RAN3 also looks random at first but when you look at it more thoroughly it seems like it's less random than the built-in random generator. Some of the points seem to be grouped together in a chain-like formation. Even though some of the points in the plot for the built-in random generator are also grouped together in short chains or clumps, the chains in the plot for the RAN3 generator are longer and seem less random because of that.

The plot for the GGL on the other hand is not random at all. All of the points are situated in lines in a regular structure. This is because GGL suffers from a short period of 2^31 - 1 steps and has problems due to correlation. First, because the next number is calculated with mod(2^31-1), there are only 2^31 - 1 different values and after that the values start repeating themselves. Second, LCGs such as GGL have a serious drawback of correlations between consecutive numbers x_i+1, x_i+2,…, x_i+d in the sequence. In d-dimensional space, the points given by these d numbers are situated in a way that forms parallel hyperplanes. These hyperplanes can be seen in the plot for the GGL generator as the two lines in the plot. The average distance between these planes, whose dimension is d – 1, is constant. The smaller the number of these planes, the less uniform is the distribution.

```python
import numpy as np

agents = 500000
init_state = 50
```

### a)

```python
a_interactions = 100 
a_states = init_state * np.ones(agents)

for i in range(a_interactions):
    a_states += np.random.choice([-1, 1], size=agents)

plt.title("PMF")
plt.hist(a_states, bins=300, density='true')
plt.show()
```

#### PMF and PDF
The probability mass function (PMF) is a binomial distribution. 

The probability density function (PDF) it approaches in the limit of infinite number of agents is a normal distribution or a Gaussian distribution with its mean at 50.

### b)

```python
b_interactions = 100 
b_states = init_state * np.ones(agents)

for i in range(b_interactions):
    b_states += np.random.choice([-0.5, 10], size=agents)

plt.title("PMF")
plt.hist(b_states, bins=300, density='true')
plt.show()
```

#### PMF and PDF
The probability mass function (PMF) is a binomial distribution. 

The probability density function (PDF) it approaches in the limit of infinite number of agents is a normal distribution or a Gaussian distribution but the mean is now different from a). The mean of this normal distribution is at around 520.

### c)

```python
c_interactions = 10
c_choices = [0.7, 1.0 / 0.7] 
c_states = init_state * np.ones(agents)

for i in range(c_interactions):
    c_states *= np.random.choice([0.7, 1.0 / 0.7], size=agents)

plt.title("PMF with linear binning")
plt.hist(c_states, bins=50, density='true')
plt.show()
   
plt.title("PMF with logarithmic binning")
plt.hist(c_states, bins=np.logspace(start=np.log10(1), stop=np.log10(250), num=50), density='true')
plt.show()
```

#### PMF
The form of the PMF resembles exponential decay.

The corresponding asymptotic PDF is lognormal distribution.
