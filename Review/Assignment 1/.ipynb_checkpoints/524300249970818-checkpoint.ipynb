## Exercise 1

```python
import matplotlib.pyplot as plt

from random import random
```

### GGL

```python
def ggl(n, seed=40000, multiplier=16807, period=2**31-1, scaling=True):

    # initialize values
    rnd_nums = [0]*n
    rnd_nums[0] = (multiplier * seed) % period

    # generate numbers based on previous numbers
    for i in range(1, n):
        rnd_nums[i] = (multiplier * rnd_nums[i-1]) % period

    if scaling:
        return [num/period for num in rnd_nums]
    else:
        return rnd_nums
```

```python
# create 500000 random numbers with GGL
rnd_nums = ggl(500000, seed=40000)

# separate even and odd indices
even_indices = rnd_nums[::2]
odd_indices = rnd_nums[1::2]

# plot
plt.subplots(figsize = (5, 5))
plt.scatter(even_indices, odd_indices, s=0.5)
plt.xlim([0.5, 0.501])

plt.xlabel('random numbers with even indices')
plt.ylabel('random numbers with odd indices')
plt.title('GGL')
plt.show()
```

We can see that the random number pairs form linear patterns meaning GGL has some predictability in its random number generation

### RAN3

```python
def ran3(n, seed=40000, m=2**32, scaling=True):
    
    rnd_nums = [0]*n
    rnd_nums[0:55] = ggl(55, seed, scaling=False)
    
    for i in range(55, n):
        rnd_nums[i] = (rnd_nums[i - 55] - rnd_nums[i - 24]) % m

    if scaling:
        return [num/m for num in rnd_nums]
    else:
        return rnd_nums
```

```python
# create 500000 random numbers with RAN3
rnd_nums = ran3(500000, seed=40000)

# separate even and odd indices
even_indices = rnd_nums[::2]
odd_indices = rnd_nums[1::2]

# plot
plt.subplots(figsize = (5, 5))
plt.scatter(even_indices, odd_indices, s=0.5)
plt.xlim([0.5, 0.501])


plt.xlabel('random numbers with even indices')
plt.ylabel('random numbers with odd indices')
plt.title('RAN3')
plt.show()
```

RAN3 creates very consistent random numbers without visible patterns

### Python built-in

```python
# create 500000 random numbers with RAN3
rnd_nums = [random() for _ in range(500000)]

# separate even and odd indices
even_indices = rnd_nums[::2]
odd_indices = rnd_nums[1::2]

# plot
plt.subplots(figsize = (5, 5))
plt.scatter(even_indices, odd_indices, s=0.5)
plt.xlim([0.5, 0.501])

plt.xlabel('random numbers with even indices')
plt.ylabel('random numbers with odd indices')
plt.title('random.random()')
plt.show()
```

random() creates very consistent random numbers without visible patterns

# Exercise 2

```python
import matplotlib.pyplot as plt
import numpy as np

from random import random
```

### a)

```python
agents = np.empty([500000, 1])

for i in range(len(agents)):
    agent = 50
    for _ in range(100):
        if random() <= 0.5:
            agent += 1
        else:
            agent -= 1
    agents[i] = agent
```

```python
plt.hist(agents, density=True, bins=50);
plt.title('a)')
plt.xlabel('Optimism score')
plt.ylabel('Probability density')
plt.show()
```

The PMF is binomial distribution and the PDF it approaches is normal distribution

### b)

```python
agents = np.empty([500000, 1])

for i in range(len(agents)):
    agent = 50
    for _ in range(100):
        if random() <= 0.5:
            agent += 10
        else:
            agent -= 0.5
    agents[i] = agent
```

```python
plt.hist(agents, density=True, bins=50);
plt.title('b)')
plt.xlabel('Optimism score')
plt.ylabel('Probability density')
plt.show()
```

The PMF is binomial distribution and the PDF it approaches is normal distribution

### c)

```python
agents = np.empty([500000, 1])

for i in range(len(agents)):
    agent = 50
    for _ in range(10):
        if random() <= 0.5:
            agent *= 0.7
        else:
            agent *= 1/0.7
    agents[i] = agent
```

```python
plt.hist(agents, density=True, bins=50);
plt.title('c) linear scale')
plt.xlabel('Optimism score')
plt.ylabel('Probability density')
plt.show()
```

The linear binning histogram looks either lognormal or exponential distribution

```python
plt.hist(np.log(agents), bins=np.logspace(start=np.log10(1), stop=np.log10(9), num=30), density=True);
plt.title('c) logarithmic scale with logaritmic binning')
plt.xlabel('Optimism score')
plt.ylabel('Probability density')
plt.show()
```

With logarithmic scale the probability density function looks approximately normal so the original PMF is likely a lognormal distribution.
The underlying process is Multiplicatively stochastic process
