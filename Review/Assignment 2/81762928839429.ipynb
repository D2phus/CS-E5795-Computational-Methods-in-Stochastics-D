```python
# As mentioned in part b) that we there only sample from by using polar coordinates and drawing r and theta from uniform distributions which would actually not return a uniform distribution over the circle as for points close to the center the distance is smaller than for points further apart. However, as this is used for a simple case we are now considering also only drawing from two uniform distributions and reject a sample if it is not inside the circle

def rejection(num):
    x = []
    y = []
    for i in range(num):
        x_pos = np.random.uniform(-1, 1)
        y_pos = np.random.uniform(-1, 1)

        if x_pos**2 + y_pos**2 >= 1:
            continue
        else:
            x.append(x_pos)
            y.append(y_pos)

    plt.figure(figsize = (5,5))
    plt.scatter(x,y, s=0.1)
    plt.title(str(num))
    plt.show()


rejection(2000)
rejection(20_000)
```

### b)

```python
def polar(num):
    x = []
    y = []
    for i in range(num):
        r = np.random.uniform()
        theta = np.random.uniform()

        x.append(r*np.cos(2*np.pi*theta))
        y.append(r*np.sin(2*np.pi*theta))

    plt.figure(figsize = (5,5))
    plt.scatter(x,y, s=0.1)
    plt.title(str(num))
    plt.show()


polar(2000)
polar(20_000)
```

What is happening here is simply the point that drawing r and $\theta$ from uniform distributions we do not end up with a uniform distribution in the whole circle. This is due to the fact that while along r and for each angle we have uniform distributions this amounts to small points which are equally likely as larger values up to 1 being closer together than the points further to the boundarz of the circle and thus not yielding a uniform distribution.

### c)

It is true that it must follow, that the corrct pdf is of the form of $p(r) = C2\pi r$ because as we need equal spacing along each different small circle.

In order to compute C we need to have the following:

$$
1 = \int_0^1 C2\pi r dr
$$

from which follows that:

$$
1 = \frac{1}{2}C2\pi 1^2
$$

Solving for C we find that it has to equal $1/\pi$

That for the cdf leaves us with $r = u^2$ Now to get the inverse we simply have to find the inverse. The first step is to swap u and r in the equation and then solve for r again which gives us $\sqrt{u}$. Thus we need to sample from the square-root of the uniform distribution.

```python
def polar(num):
    x = []
    y = []
    for i in range(num):
        r = np.sqrt(np.random.uniform())
        theta = np.random.uniform()

        x.append(r*np.cos(2*np.pi*theta))
        y.append(r*np.sin(2*np.pi*theta))

    plt.figure(figsize = (5,5))
    plt.scatter(x,y, s=0.1)
    plt.title(str(num))
    plt.show()


polar(2000)
polar(20_000)
```

This looks way better than what we had in b).

To determine the C in this equation for the support $x \in [1,\infty)$ we have to take the integral of $f(x)$ as the area under the pdf has to equal 1. Therefore we get:

$$
1 = \int_1^{\infty} Cx^{-\frac{3}{2}}dx
$$

from which follows

$$
1 = [-2Cx^{-\frac{1}{2}}]_0^{\infty} = 0 - (-2C) = 2C
$$

The first term drops to 0 as $1/x^{0.5}$ tends to 0 for x approaching infinity. Now we can see that $C=0.5$

\begin{align*}
x &= 0.5y^{-0.5} \\
y &= 0.5x^{-0.5} \\
2y &= x^{-0.5} \\
2y &= \frac{1}{x^{0.5}} \\
x^{0.5} &= \frac{1}{2y} \\
x &= \left(\frac{1}{2y}\right)^2 = \frac{1}{4y^2}
\end{align*}
