```python
import math 
import time 

import numpy as np 
import matplotlib.pyplot as plt
```

# Gibbs sampler 
We believe the measured values are normally distributed and i.i.d. 
$$
\begin{aligned}
x_i &\sim N(\mu, \frac{1}{\tau})
\end{aligned}
$$
then the **likelihood** for $n=20$ observations $\boldsymbol{x}= (x_1, \ldots, x_n)^T$
$$
\begin{aligned}
\pi(\boldsymbol{x}|\mu, \tau) & = \prod ^n _{i=1}\sqrt{\frac{\tau}{2\pi} }e^{-\frac{\tau}{2}(x_i-\mu)^2}
\end{aligned}
$$

The posterior distribution $\pi(\mu, \tau|x)$, full conditionals $\tau |\mu, x, \mu|\tau, x$ are also given. 

## 0. Gibbs sampler class 
First Let's implement the Gibbs sampler class! There are some details to which we should pay attention:

1. `np.random.gamma(shape, scale)`, where the scale parameter is the inverse of beta!
2. burn-in period (time to converge): some initial values actually should be discarded. The burn-in period is checked in [Appendix](#Appendix). We can see the MCMC for both $\mu$ and $\sigma$ converges in the first $300$ steps, i.e. the burn-in period is no more than $300$ steps long. We can just neglect it.

>  Once initialised, the
sampler proceeds with alternate simulations from the full conditional distributions.
The first few (hundred?) values should be discarded, and the rest can give information
about the joint posterior distribution and marginals.
</br>...</br>
the first portion (say, a third) of any run should be discarded as ‘burn-in’

2. to remedy the irreversibility issue, we scan the components first in order and then reversed order
3. 'online' update: new values are used to update following variants in the same round.

```python
class GibbsSampler:
    """Gibbs sampler algorithm to simulate the multivariate joint distribution, 
    where there variants are not independent while full conditionals are known. """
    def __init__(self, 
                 n: int=20, 
                 a: float=1, 
                 b: float=1, 
                 c: float=3, 
                 d: float=1/100, 
                 s2: float=20, 
                 e: float=30): 
        """
        Args: 
        ----
        e: average of measurements, i.e. point estimate
        """
        self.n = n 
        self.a = a
        self.b = b 
        self.c = c 
        self.d = d
        
        self.s2 = s2
        self.s = math.sqrt(self.s2)
        self.e = e 
        
    @property
    def prec_gamma_shape(self):
        """shape parameter for the Gamma-shape full conditional for precision"""
        return self.a + self.n/2
    
    def prec_gamma_scale(self, mu): 
        """scale parameter for the Gamma-shape full conditional for precision"""
        beta = self.b + 0.5*((self.n-1)*self.s2 + self.n*(self.e-mu)**2)
        scale = 1/beta # NOTE that scale is the inverse of beta 
        return scale
    
    def mu_normal_var(self, prec): 
        """variance parameter for the Normal-shape full conditional for mu"""
        return 1/(self.n*prec+self.d)
    
    def mu_normal_mean(self, prec):
        """mean parameter for the Normal-shape full conditional for mu"""
        mean = self.c*self.d + self.n*prec*self.e
        mean /= self.n*prec + self.d
        return mean 
    
    def sampling_prec_full_cond(self, mu: float):
        """sampling from the precision full conditional"""
        shape = self.prec_gamma_shape
        scale = self.prec_gamma_scale(mu)
        s = np.random.gamma(shape, scale, 1)[0]
        return s
    
    def sampling_mean_full_cond(self, prec: float):
        """sampling from mean full conditional"""
        mean = self.mu_normal_mean(prec)
        var = self.mu_normal_var(prec)
        sigma = math.sqrt(var)
        s = np.random.normal(mean, sigma, 1)[0]
        return s
    
    def __call__(self, num_samples: int=100000):
        """sampling mu and precision from the simulated joint posterior distribution for `num_samples` times.
        """
        # prior means for mean and precision
        # NOTE that initial values should not be taken into account!
        mu = self.a / self.c 
        precision = self.c
        
        mu_list = list()
        prec_list = list()
        
        st = time.time()
        for idx in range(num_samples): 
        # to remedy the irreversibility issue, we scan the components first in order and then reversed order
            if idx % 2 == 0:
                mu = self.sampling_mean_full_cond(precision)
                precision = self.sampling_prec_full_cond(mu)
            else:
                precision = self.sampling_prec_full_cond(mu)
                mu = self.sampling_mean_full_cond(precision)
                
            mu_list.append(mu)
            prec_list.append(precision)
        
        mu_list = np.asarray(mu_list)
        prec_list = np.asarray(prec_list)
        et = time.time()
        duration = et-st
        print(f'Sampling finished. use time: {duration: .3f}s')
        return mu_list, prec_list
```

```python
gibbs_sampler = GibbsSampler()
mu, precision = gibbs_sampler()
sigma = 1/np.sqrt(precision)
```

## 1. Plotting function
Then let's implement the plotting functions. 

```python
def plot_joint_posterior_distribution(mu, precision, axs=None, trace=False):
    if axs is None: 
        fig, axs = plt.subplots()
    
    axs.set_title('Joint posterior')
    axs.set_xlabel('mu')
    axs.set_ylabel('precision')
    if trace:
        
        axs.plot(mu, precision, '-')
    else: 
        axs.scatter(mu, precision, alpha=0.3, c='orange')
    return axs 
```

```python
def plot_marginal_trace(data, axs=None):
    if axs is None: 
        fig, axs = plt.subplots()
    
    axs.set_xlabel('Iteration')
    axs.set_ylabel('Value')
    
    num_samples = len(data)
    axs.plot([*range(num_samples)], data)
    return axs
```

```python
def plot_marginal_distribution(data, axs=None):
    if axs is None: 
        fig, axs = plt.subplots()
    
    axs.set_xlabel('Value')
    axs.set_ylabel('Frequency')
    
    values, edges, _ = axs.hist(data, bins=30, density=True)
    return axs
```

## 2. Plots
Finally, the plots are shown below. 

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
plot_joint_posterior_distribution(mu, precision, axs=axs[0], trace=False)
plot_joint_posterior_distribution(mu, precision, axs=axs[1], trace=True)
fig.tight_layout()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(14, 3))
axs[0].set_title('Marginal for mu')
plot_marginal_trace(mu, axs=axs[0])
axs[1].set_title('Marginal for precision')
plot_marginal_trace(precision, axs=axs[1])
axs[2].set_title('Marginal for sigma')
plot_marginal_trace(sigma, axs=axs[2])

fig.tight_layout()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(14, 3))
axs[0].set_title('Marginal for mu')
plot_marginal_distribution(mu, axs=axs[0])
axs[1].set_title('Marginal for precision')
plot_marginal_distribution(precision, axs=axs[1])
axs[2].set_title('Marginal for sigma')
plot_marginal_distribution(sigma, axs=axs[2])

fig.tight_layout()
```

## Appendix: check burn-in period aka time to converge (visually)
We check for the build-in period by looking over $100$ steps starting from $0, 100, 200$ of several markov chains and see if the chains are overlapping. 

It looks like MCMC already converges at least in the first $300$ steps, i.e. the burn-in period is even shorter than $300$ steps, so we can just neglect it.

There is another way (not shown here) to do convergence check: $\hat{R}$ (potential scale reduction factor). 

```python
initial_steps = [100, 200, 300]
span = 100
fig, axs = plt.subplots(len(initial_steps), 2)
timesteps = np.asarray([*range(100000)])

for j in range(3): 
    gibbs_sampler = GibbsSampler()
    mu, precision = gibbs_sampler()
    sigma = 1/np.sqrt(precision)
    for i, istep in enumerate(initial_steps):
        axs[i][0].set_title('mu')
        axs[i][0].plot(timesteps[(istep-span):istep], mu[(istep-span):istep])
        axs[i][1].set_title('sigma')
        axs[i][1].plot(timesteps[(istep-span):istep], sigma[(istep-span):istep])

fig.tight_layout()
```
